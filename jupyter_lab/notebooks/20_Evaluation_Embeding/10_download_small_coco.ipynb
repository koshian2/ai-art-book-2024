```python
import datasets
import imagehash
from collections import defaultdict
import random
from PIL import Image
import os

# データセットを読み込み
dataset = datasets.load_dataset('RIW/small-coco', split='train')

hash_dict = defaultdict(list)

# データセットの画像URLを取得し、ImageHashを計算
image_hashes = []
for idx, entry in enumerate(dataset):
    image_data = entry['image']
    image_hash = imagehash.phash(image_data, hash_size=8)
    image_hashes.append(image_hash)
    hash_dict[str(image_hash)].append(idx)

# 結果を表示
# for hash_value, indices in hash_dict.items():
#     print(f"Hash: {hash_value} -> Indices: {indices}")

# 乱数のシードを固定
random.seed(42)

# ランダムに100個のハッシュ値を選択
random_hashes = random.sample(list(hash_dict.keys()), 100)

# ランダムに選択された100個のハッシュ値からランダムなインデックスを選び、画像を保存
for i, hash_value in enumerate(random_hashes):
    random_index = random.choice(hash_dict[hash_value])
    image_data = dataset[random_index]['image']
    caption_data = dataset[random_index]['caption']
    os.makedirs("small_coco", exist_ok=True)
    image_data.save(f'small_coco/random_image_{i:03}.png')
    with open(f'small_coco/random_caption_{i:03}.txt', "w", encoding="utf-8") as fp:
        fp.write(caption_data)

print("Images saved successfully.")
```
